```python
import bs4
import requests
```

```python
url="https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response=requests.get(url)
```

```python
response
```

```python
response.content
```

```python
type(response.content)
```

```python
# HTML parser
soup=bs4.BeautifulSoup(response.content)
```

```python
soup
```

```python
 element=soup.find_all('img')
```

```python
element
```

```python
element[3]
```

```python
element[3].attrs
```

```python
element[3].attrs['src']
```

```python
with open('motivation.jpg','wb') as f:
    img_url=element[3].attrs['src']
    response=requests.get(img_url)
    f.write(response.content)
```

```python
element[36]
```

```python
for i in range(3,36):
    print(f'extracting {i} image',end='\r')
    url=element[i].attrs['src']
    name=f'motivation{i}.jpg'
    with open(name,'wb')as f:
        response=requests.get(url)
        f.write(response.content)
print('------------------------------done------------------------',end='\r')
```
